```python
import numpy as np

#func for one hot encoding
def convert_to_one_hot(data_index , vocab_size):
  vec = np.zeros(vocab_size)
  vec[data_index] = 1
  return vec

x_train = []
y_train = []


#appending context words in x_train and y_train lists
for data_word in data:
  print(data_word)
  x_train.append(convert_to_one_hot(word_int[data_word[0]],vocab_size))
  y_train.append(convert_to_one_hot(word_int[data_word[1]],vocab_size))


#creating arrays 
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)

print(x_train[0])
print(y_train[0])  
```

```python
# MAKING x_train and y_train
# -------------> A placeholder is simply a variable that we will assign data to at a later time. 


x_input = tf.placeholder(tf.float32,shape=(None,vocab_size))
y_label = tf.placeholder(tf.float32,shape=(None,vocab_size))
```

```python
DIM = 20 #dimension

# CREATING OUR NEURAL NETWORK: W1, v1, W2, v2
W1 = tf.Variable(tf.random_normal([vocab_size,DIM]))
v1 = tf.Variable(tf.random_normal([DIM]))
hidden_layer = tf.add(tf.matmul(x_input,W1),v1)


W2 = tf.Variable(tf.random_normal([DIM,vocab_size]))
v2 = tf.Variable(tf.random_normal([vocab_size]))
prediction = tf.nn.softmax(tf.add(tf.matmul(hidden_layer,W2),v2))
```

```python
# CREATING OUR NEURAL NETWORK: LOSS FUNCTION, LEARNING ALGO (GD)

sess = tf.Session()
init = tf.global_variables_initializer()       
sess.run(init)                                                         

cross_entropy_loss = tf.reduce_mean(-tf.reduce_sum(y_train * tf.log(prediction), reduction_indices=[1]))  # The loss function

train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy_loss)
```

```python
# TRAINING

n = 4000 #iterations           

for _ in range(n):
    sess.run(train_step, feed_dict={x_input: x_train, y_label: y_train})
    print('loss is : ', sess.run(cross_entropy_loss, feed_dict={x_input: x_train, y_label: y_train}))
```
